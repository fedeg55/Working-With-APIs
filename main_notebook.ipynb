# Use this notebook for your 3 choices of APIs

```python
import requests
import json
import pandas as pd
from pandas.io.json import json_normalize

response = requests.get('https://dog.ceo/api/breeds/list/all')
response
results = response.json()
results
```

```python
len(results['message'])
```

```python
import pandas as pd

data = {
    'breed': [],
    'sub_breed': []
}

for breed, sub_breeds in results['message'].items():
    for sub_breed in sub_breeds:
        data['breed'].append(breed)
        data['sub_breed'].append(sub_breed)

df = pd.DataFrame(data)
df
```

```python
data_list = [{'Breed': key, 'SubBreed': ', '.join(value) if value else '0'} for key, value in results['message'].items()]
data_list
d = pd.DataFrame.from_dict(data_list)
d.head(20)
```
